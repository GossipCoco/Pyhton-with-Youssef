```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], kde=True)
plt.title('Histogramme de la Colonne Budget')
plt.xlabel('Valeur de chaque ventes')
plt.ylabel('Budget en Euro')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], kde=True, color='pink')
plt.title('Histogramme de la Colonne Budget')
plt.xlabel('Valeur de chaque vente')
plt.ylabel('Budget en Euro')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], kde=False, color='pink')
plt.title('Histogramme de la Colonne Budget')
plt.xlabel('Valeur de chaque vente')
plt.ylabel('Budget en Euro')
plt.grid(True)  # Ajout du quadrillage
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.histplot(df['Budget'], kde=True, color='#FF6998', binwidth=250)  # Ajustez binwidth selon vos données
plt.title('Histogramme de la Colonne Budget')
plt.xlabel('Valeur de chaque vente')
plt.ylabel('Budget en Euro')
plt.grid(True)  # Ajout du quadrillage
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='IdProduit', y='Budget', data=df, color='#FF6998')
plt.title('Boxplot du Budget par IdProduit vendu')
plt.xlabel('IdProduit vendu')
plt.ylabel('Budget')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Remise', y='Budget', data=df)
plt.title('Scatter Plot de Colonne Budget   vs Remise   ')
plt.xlabel('Colonne Remise')
plt.ylabel('Colonne Budget')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Remise', y='Budget', data=df)
plt.title('Scatter Plot de Colonne Budget   vs Remise   ')
plt.xlabel('Colonne Remise')
plt.ylabel('Colonne Budget')
plt.show()
```

```python
# Exemple 4: Heatmap de corrélation

#Sélectionner les colonnes numériques
numeric_cols = df.select_dtypes(include=['number']).columns
df_numeric = df[numeric_cols]

#Heatmap de corrélation
plt.figure(figsize=(12, 8))
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap de Corrélation')
plt.show()
```
